```python
import nltk
from nltk.tokenize import word_tokenize
import re
```

```python
tokens=[]
for i in range(len(data)):
    s=data[i]
    s=re.sub(r'http\S+', ' ', s)
    s=re.sub(r'pic.twitter.com\S+',' ',s)
    s=re.sub(r'@\S+', ' ',s)
    s=re.sub(r'[^\w\s]',' ',s)
    s=s.lower()
    if "ðŸ" in s:
        #print("ðŸ")
        s=re.sub("ðŸ", " ",s)
        #print(s.split())
    s=re.sub(r'\d', ' ', s)
    l=word_tokenize(s)
    tokens.extend(l)
    data[i]=s
```
